# NORMALIZATION

```python
import pandas as pd
s = pd.Series([1000,2000,3000,5000,9000])
s
```

```python
s.min()
```

```python
s.max()
```

```python
s_norm = (s - s.min())/(s.max() - s.min())
print(s_norm)
```

# IMPORT MODULES

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
s_norm = MinMaxScaler().fit_transform(s.values.reshape(-1, 1))
print(s_norm)
```

```python
df = pd.DataFrame({
    "Temp":[20,25,30,35,40],
    "Humidity":[30,45,60,70,75]
})
df
```

```python
norm_data = MinMaxScaler().fit_transform(df)
norm_data
```

```python
norm_df = pd.DataFrame(norm_data, columns=df.columns)
norm_df
```

# STANDARDIZATION:(Z_SCORE)

```python
import pandas as pd
s = pd.Series([1000,2000,3000,5000,9000])
s
```

```python
s.mean()
```

```python
s.std()
```

```python
Z_Score=(s - s.mean()) / s.std()
print(Z_Score)
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
Z_Score = StandardScaler().fit_transform(s.values.reshape(-1, 1))
print(Z_Score)
```

```python
Z_Score = (s - s.mean()) / s.std(ddof=0)
print(Z_Score)
```

```python
df = pd.DataFrame({
    "Temp":[20,25,30,35,40],
    "Humidity":[30,45,60,70,75]
})
df
```

```python
Z_score=StandardScaler().fit_transform(df)
Z_score_df=pd.DataFrame(Z_score,columns=df.columns)

print(Z_score_df)
```
