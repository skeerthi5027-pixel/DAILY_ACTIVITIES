# Outliers

```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "Marks":[1,2,3,4,5,6,7,8,20]
})
```

```python
data
```

```python
data.describe()
```

# IQR Method

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
LB = Q1 - 1.5 * IQR
UB = Q3 - 1.5* IQR
outliers = data[(data < (LB)) | (data > (UB))]
```

# Z-Score

```python
data = pd.DataFrame({
    "Study_hours":[1,2,3,4,5,6,7,8,20]
})
```

```python
data
```

```python
z_scores = np.abs(stats.zscore(data["Study_hours"]))
```

```python
outliers = data[z_scores > 2]
```

```python
outliers
```

# 

```python
plt.scatter(range(len(data)), data["Study_hours"])
plt.xlabel("Index")
plt.ylabel("Study hours")
plt.title("Scatter Plot showing Outlier")
plt.show()
```

```python
plt.hist(data["Study_hours"], 6)
plt.xlabel("Study hours")
plt.ylabel("Frequency")
plt.title("Histogram showing Outlier Effect")
plt.show()
```
